```python
import cv2
```

```python
import matplotlib.pyplot as plt
```

```python
config_file = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model = 'frozen_inference_graph.pb'
```

```python
model = cv2.dnn_DetectionModel(frozen_model, config_file)
```

```python
classLabels = []
file_name = 'coco.names'
with open(file_name, 'rt') as fpt:
    classLabels = fpt.read().rstrip('\n').split('\n')
```

```python
print(len(classLabels))
print(classLabels)
```

```python
model.setInputSize(320,320)
model.setInputScale(1.0/127.5)  ## 255/2 = 127.5
model.setInputMean((127.5,127.5,127.5))  ## mobilenet => [-1,1]
model.setInputSwapRB(True)
```

# read an image

```python
img = cv2.imread('man_with_car.jpg')
```

```python
plt.imshow(img) ## bgr
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) ## rgb
```

```python
ClassIndex, confidence, bbox = model.detect(img, confThreshold=0.5)
```

```python
print(ClassIndex)
```

```python
font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN
for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
    cv2.rectangle(img, boxes, (255,0,0), 2)
    cv2.putText(img, classLabels[ClassInd-1], (boxes[0]+10, boxes[1]+40), font, fontScale=font_scale, color=(0,255,0), thickness=3)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

# Video Demo

```python
video_file = 'Walking in the city.mp4'
#video_file = 'New York City traffic.mp4'
#video_file = 0

cap = cv2.VideoCapture(video_file)

if not cap.isOpened():
    cap = cv2.VideoCapture(0)
if not cap.isOpened():
    raise IOError('Cannot open video')
    

font_scale = 3
font = cv2.FONT_HERSHEY_PLAIN

while True:
    ret, frame = cap.read()
    if (ret != True):
        break
        
    ClassIndex, confidence, bbox = model.detect(frame, confThreshold=0.50)
    
    print(ClassIndex)
    if (len(ClassIndex)!=0):
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
            if (ClassInd <= 80):
                cv2.rectangle(frame, boxes, (255,0,0), 2)
                label = classLabels[ClassInd-1] + ' : ' + str(round(conf*100,1)) + '%'
                cv2.putText(frame, label, (boxes[0]+10, boxes[1]+40), 
                            font, fontScale=font_scale, color=(0,255,0), thickness=3)
                
    cv2.imshow('Object Detection Tutorial', frame)
    
    if cv2.waitKey(2) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
    
```
